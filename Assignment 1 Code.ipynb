```python
#importing packages

import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.util import random_noise
from skimage import feature
import cv2
import glob
```

```python
#resizing the images, prior to running them through edge detection
from PIL import Image

for filename in sorted(glob.glob("Outputs_png/*.png")):
    img = Image.open(filename)
    img = img.resize((512, 512))
    img.save(filename)
```

```python
#defining an edge detection function - adjusted from stack overflow (documentation in ReadMe)
def auto_canny(image, sigma=0.33):
    # Compute the median of the single channel pixel intensities
    v = np.median(image)

    # Apply automatic Canny edge detection using the computed median
    #the saved images will be saved at our inputs_png
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    return cv2.Canny(image, lower, upper)
```

```python
# Read in each image and convert to grayscale
images = [cv2.imread(file,0) for file in sorted(glob.glob('Outputs_png/*.png'))] 
print(images)
# Iterate through each image, perform edge detection, and save image
number = 1
for image in images:
    canny = auto_canny(image)
    cv2.imwrite('{}.png'.format(number), canny)
    number += 1
```

This dataset was created in part to use on my Mini-Project assignment for AI-4-Media course. For the Pix2Pix model to work, we need to save the inputs and outputs side by side as one image. The Pix2Pix model is set up to separate these images as tensors and use the paired tensors to train the model. As my images are saved separately, the below code will combine them so the Pix2Pix model to work. The code I am using for this has been adapted from this code found on stack overflow: https://stackoverflow.com/questions/65342361/photos-side-by-side. 

```python
#import packages
from PIL import Image, ImageOps, ImageDraw
```

```python
output_filenames = sorted(glob.glob('Outputs_png/*.png')) 
input_filenames = sorted(glob.glob('Inputs_png/*.png'))
i = 0 
for output_file in output_filenames:
    output_file = output_file.split('/')
    output_file = output_file[1] 
    for input_file in input_filenames:
        input_file = input_file.split('/')
        input_file = input_file[1]
        if output_file == input_file:
            output_image = Image.open('Outputs_png/' + output_file)
            input_image = Image.open('Inputs_png/' + input_file)  
            both = Image.new('RGB', (1024,512))

            both.paste(output_image)
            both.paste(input_image, (512,0))
            i += 1
            both.save(str(i) + '.png')
            
```
